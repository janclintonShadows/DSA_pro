```python
#importando os modulos importantes:
import pandas as pd
import matplotlib.pyplot as plt
#importar a biblioteca para usar o outlook "pip install pywin32"
import win32com.client as win32
```

```python
#importando o nosso banco de dados ou arquivo
df = pd.read_excel("Vendas.xlsx")
```

```python
#visualizando os primeiros elementos do banco de dados
df.head()
```

```python
#faturamento por loja
faturamento = df[['ID Loja','Valor Final']].groupby('ID Loja').sum()
faturamento
```

```python
#Quantidade de produtos vendidos por loja
quantidade = df[['ID Loja','Quantidade']].groupby('ID Loja').sum()
quantidade
```

```python
#ticket médio por produtos em cada loja
#usamos o comando to_frame() para tornar a divisão em uma tabela
ticket_medio = (faturamento['Valor Final']/quantidade['Quantidade']).to_frame('Ticket Medio')
ticket_medio
```

```python
#Enviar um email com relatorio -> Esse metodo é para o Outlook

#Conectando com o Outlook no computador
outlook = win32.Dispatch('outlook.application')
#Cria um Email
mail = outlook.CreateItem(0)
#Email de Destino
mail.To = "janclinton.brightminds@gmail.com"
#Assunto do Email
mail.Subject = 'Enviando Email com o Python'
#Corpo do Email
mail.HTMLBody = f'''
<p>Segue o conteudo do nosso email:</p>
<br>
Tabela de Faturamento<br>
{faturamento.to_html(formatters={'Valor Final':'R${:,.2f}'.format})}
<br>
Tabela de Quantidade<br>
{quantidade.to_html()}
<br>
Tabela de Ticket Médio<br>
{ticket_medio.to_html(formatters={'Ticket Medio':'R${:,.2f}'.format})}
'''
#Enviar o Email
mail.Send()
```
